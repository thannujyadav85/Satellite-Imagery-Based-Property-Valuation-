```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_excel("data/train(1).xlsx")
```

```python
plt.figure(figsize=(6,4))
sns.histplot(train_df['price'], bins=50, kde=True)
plt.title("House Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x=train_df['sqft_living'],
    y=train_df['price'],
    alpha=0.3
)
plt.title("Price vs Living Area")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=train_df['waterfront'], y=train_df['price'])
plt.title("Waterfront vs Price")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=train_df['grade'], y=train_df['price'])
plt.title("Grade vs Price")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(
    train_df['long'],
    train_df['lat'],
    c=train_df['price'],
    cmap='viridis',
    s=5
)
plt.colorbar(label="Price")
plt.title("Geographical Distribution of House Prices")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(
    train_df['long'],
    train_df['lat'],
    c=train_df['waterfront'],
    cmap='coolwarm',
    s=5
)
plt.colorbar(label="Waterfront")
plt.title("Waterfront Properties Distribution")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
